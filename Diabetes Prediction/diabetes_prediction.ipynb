# 1.3. About Dataset &#x1F4C4;:

1. gender: represents male, female or others

2. age: represnts the ages of the diabetic and non diabetic patients.

3. hypertension: represents weather the patient has hypertension or not.

4. heart_disease: represemts weather the patient has heart disease or not.

5. smoking_history: represents the smoking history among patients (i.e. current, never. no info, former, not current, ever).

6. bmi: represents the body mass index of the patient.

7. HbA1c_level: represents the hemoglobin level of the patient.

8. blood_glucose_level: Blood glucose level refers to the amount of glucose in the bloodstream at a given time.

9. diabetes: weather the patient is diabetic or not.

```python
X_train=df1.drop(df1[['diabetes']], axis=1)
y_train=df1[['diabetes']]
```

```python
y_train
```

```python
print('X_train Shape:', X_train.shape)
print('X_test Shape:', X_test.shape)
print('y_train Shape:', y_train.shape)
print('y_test Shape:', y_test.shape)
```

# **<center> **6. MODEL EVALUATION :pencil::mag:** </CENTER>**

```python
X_train, X_test, y_train, y_test=train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

# :arrow_right: 6.1. Random Forest Classifier:

```python
rfc=RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
y_pred=rfc.predict(X_test)
y_pred
```

```python
# accuracy score
print('Accuracy Score:', accuracy_score(y_test, y_pred))
```

```python
# confusion matrix
cm=confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# classification report
report=classification_report(y_test, y_pred)
print('Classification Report:')
print(report)
```

# :arrow_right: 6.2. K-Nearest Neighbour:

```python
knn=KNeighborsClassifier()

knn.fit(X_train, y_train)
```

```python
y_pred=knn.predict(X_test)
y_pred
```

```python
print('Accuracy Score:', accuracy_score(y_test, y_pred))
```

```python
# confusion matrix
cm=confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=['Class 0', 'Class 1', 'Class 2'], yticklabels=['Class 0', 'Class 1', 'Class 2'])
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# classification report
report=classification_report(y_test, y_pred)
print('Classification Report:')
print(report)
```
